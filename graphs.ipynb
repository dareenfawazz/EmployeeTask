```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_excel("EmployeeAnalysis.xlsx")
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(data=df, x='department', y='age')
plt.title("Age Distribution by Department")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.scatterplot(data=df, x='age', y='annual_salary', hue='department')
plt.title("Salary vs Age by Department")
plt.show()
```

```python
genderDept = df.groupby(['department', 'gender']).size().unstack().fillna(0)
genderDept.plot(kind='bar', stacked=True, figsize=(6, 6))
plt.title("Gender Distribution by Department")
plt.ylabel("No of Employees")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.histplot(df['bonus_%'], bins=30, kde=True, color='Green')
plt.title("Bonus Distribution")
plt.xlabel("Bonus")
plt.show()
```
